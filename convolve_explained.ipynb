```python
from scipy import misc
```

```python
img = misc.ascent()
print(img)
print(img.shape)
```

```python
import matplotlib.pyplot as plt
import numpy
```

```python
plt.imshow(img)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.shape
```

```python
kernel = numpy.array([
    [0, 0, 0],
    [0, -4, 0],
    [0, 0, 0],
])
```

```python
from scipy.signal import convolve2d
```

```python
img1 = convolve2d(img, kernel, boundary='symm')
```

```python
img1
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img1, cmap='gray')
```
